```python
import pandas as pd
import requests

# Función para obtener países por moneda
def get_countries_by_currency(currency_code):
    response = requests.get(f"https://restcountries.com/v3.1/currency/{currency_code}")
    if response.ok:
        countries = response.json()
        # Prepare the data for DataFrame
        data = []
        for country in countries:
            country_data = {
                'Official Country Name': country['name']['official'],
                'Capital': country.get('capital', ['No Capital'])[0],
                'Google Maps URL': country['maps']['googleMaps']
            }
            data.append(country_data)
        return pd.DataFrame(data)
    else:
        return pd.DataFrame()


# Función para obtener países por idioma
def get_countries_by_language(languages):
    countries_info = []
    for language in languages:
        response = requests.get(f"https://restcountries.com/v3.1/lang/{language}")
        if response.ok:
            for country in response.json():
                countries_info.append({
                    'Language': language,
                    'Country': country['name']['common'],
                    'Population': country['population'],
                    'Gini Index': country.get('gini', {}).get('2019', 'No Data')
                })
    return pd.DataFrame(countries_info)

# Obtiene los países que usan el Euro
euro_countries_df = get_countries_by_currency('EUR')
print(euro_countries_df)
euro_countries_df.to_csv('euro_countries.csv', index=False)

# Obtiene los países que hablan español, inglés y francés
languages_df = get_countries_by_language(['spanish', 'english', 'french'])
print(languages_df)
languages_df.to_csv('language_countries.csv', index=False)
```
