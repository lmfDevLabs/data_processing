Creación de una columna de identificación y creación de la variable *Borb (bulliesh or beariesh)* para cada jornada y simbolo

```python
import pandas as pd

# Diccionario con los nombres de las compañías y sus correspondientes DataFrames
companies_data = {
    'AAPL': df_cleanning_aapl,
    'MSFT': df_cleanning_msft,
    'AMZN': df_cleanning_amzn,
    'NVDA': df_cleanning_nvda,
    'GOOGL': df_cleanning_googl
}

# Bucle para procesar cada DataFrame
for symbol, df in companies_data.items():
    # Asegúrate de que 'Date' sea una columna y no el índice
    df.reset_index(inplace=True)
    # Agregar una columna de identificación
    df['Company'] = symbol

# Concatenar todos los DataFrames en uno solo
df_combined = pd.concat(companies_data.values())

# Reordenar las columnas para que 'Company' y 'Date' sean las primeras
cols = ['Company', 'Date'] + [col for col in df_combined.columns if col not in ['Company', 'Date']]
df_combined = df_combined[cols]

# Crear la variable "Borb" para cada acción en el DataFrame combinado
for symbol in ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL']:
    df_combined.loc[df_combined['Company'] == symbol, 'Borb'] = (df_combined.loc[df_combined['Company'] == symbol, 'Close'] > df_combined.loc[df_combined['Company'] == symbol, 'Close'].shift(1)).astype(int)

# Asegurarse de que 'Date' sea una columna y no el índice
df_combined.reset_index(inplace=True, drop=True)

# Eliminar la primera fila para cada acción ya que no tiene un día anterior con el cual comparar
df_combined = df_combined.groupby('Company').apply(lambda x: x.iloc[1:]).reset_index(drop=True)

# Mostrar las primeras filas de df_combined para verificar
print(df_combined.head())
```

Definición y comprobación de coincidencia en el time frame, tanto del indice como de los simbolos. Ademas de creacion de variable *InBorb* para rastrear subidas o bajadas en el indice y Borb para las acciones o simbolos

```python
import pandas as pd

# Cargar los datos
dataStocks = pd.read_csv('combined_stock_data.csv')
sp500_data = pd.read_csv('sp500_data.csv')

# Convertir la fecha si es necesario
dataStocks['Date'] = pd.to_datetime(dataStocks['Date'])
sp500_data['Date'] = pd.to_datetime(sp500_data['Date'])

# Ordenar los datos por fecha
sp500_data.sort_values(by='Date', inplace=True)
dataStocks.sort_values(by='Date', inplace=True)

# Calcular la variable de subida o bajada para el S&P 500
sp500_data['InBorb'] = (sp500_data['Close'] > sp500_data['Close'].shift(1)).astype(int)

# Alinear los datos con las acciones
aligned_data = pd.merge(dataStocks, sp500_data, on='Date', suffixes=('_stock', '_sp500'))

# Eliminar la primera fila de sp500_data ya que no tiene un día anterior con el cual comparar
# (esto es solo si necesitas que la fila esté también en el DataFrame fusionado)
sp500_data = sp500_data.iloc[1:]

# Alinear nuevamente después de eliminar la primera fila de sp500_data si es necesario
aligned_data = pd.merge(dataStocks, sp500_data, on='Date', suffixes=('_stock', '_sp500'))

# Ahora calcular la variable de subida o bajada para cada acción
aligned_data['Borb'] = (aligned_data['Close_stock'] > aligned_data['Close_stock'].shift(1)).astype(int)

# En este punto, deberías tener dos columnas 'InBorb' y 'Borb'
# 'InBorb' indica si el S&P 500 subió o bajó
# 'Borb' indica si la acción específica subió o bajó

# Mostrar las primeras filas de los datos alineados para verificar
print(aligned_data.head())
```

# Ingenieria de caracteristicas para simbolos

# Cambios Porcentuales

Relevancia Económica: Los cambios porcentuales en los precios de las acciones proporcionan una medida relativa de cómo una acción se está desempeñando, independientemente de su precio absoluto. Esto permite comparar el rendimiento relativo de diferentes acciones.
Normalización: Al convertir los precios en cambios porcentuales, normalizamos los datos, lo que es útil para comparar acciones con diferentes rangos de precios.

# Medias Móviles

Tendencias: Las medias móviles suavizan las fluctuaciones a corto plazo y resaltan las tendencias a largo plazo en los datos. Son útiles para identificar tendencias alcistas o bajistas en el precio de una acción.
Señales de Trading: En el análisis técnico, las medias móviles se utilizan a menudo para generar señales de compra o venta. Por ejemplo, cuando una media móvil a corto plazo cruza por encima de una media móvil a largo plazo, se considera una señal alcista.

```python
# Calcular el cambio porcentual diario en los precios de cierre de las acciones
for company in ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL']:
    aligned_data[f'{company}_pct_change'] = aligned_data.query(f"Company == '{company}'")['Close_stock'].pct_change()

# Calcular medias móviles simples (SMA) para diferentes períodos de tiempo
moving_averages = [5, 10, 20]
for ma in moving_averages:
    for company in ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL']:
        aligned_data.loc[aligned_data['Company'] == company, f'SMA{ma}_{company}'] = aligned_data.query(f"Company == '{company}'")['Close_stock'].rolling(window=ma).mean()

# Inspeccionar los resultados
print(aligned_data)
```

Estimación de promedios para la variable *Borb* (simbolos) y creación de columna *Daily_Borb_Average* con este dato, ademas de adhesion de datos de InBorb (indice)

```python
import pandas as pd

# Asumimos que 'dataStocks' y 'sp500_data' ya están cargados y preprocesados como se discutió anteriormente
dataResumen = pd.read_csv('combined_stock_data_features.csv')

# Ordenar los datos por fecha si aún no lo están
dataResumen.sort_values(by='Date', inplace=True)

# Calcular la variable de subida o bajada para cada acción y para el S&P 500
dataResumen['Borb'] = (dataResumen['Close_stock'] > dataResumen['Close_stock'].shift(1)).astype(int)
dataResumen['InBorb'] = (dataResumen['Close_sp500'] > dataResumen['Close_sp500'].shift(1)).astype(int)

# Agrupar por fecha y calcular el promedio de 'Borb' para las acciones por día
daily_borb_average = dataResumen.groupby('Date')['Borb'].mean().reset_index(name='Daily_Borb_Average')

# Asegurarse de que solo tomamos los valores de 'InBorb' desde el segundo día (ya que el primer día no tiene un día anterior para comparar)
sp500_inborb = dataResumen[['Date', 'InBorb']].iloc[1:]

# Fusionar ambos DataFrames en uno nuevo basado en la fecha
merged_data = pd.merge(daily_borb_average, sp500_inborb, on='Date')

# Mostrar las primeras filas del DataFrame fusionado para verificar
print(merged_data.head())
```

analisis trimestral

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Cargar los datos
poder_data = pd.read_csv('poder.csv')

# Convertir la columna 'Date' a datetime
poder_data['Date'] = pd.to_datetime(poder_data['Date'])

# Asegurarse de que los datos estén ordenados por fecha
poder_data.sort_values('Date', inplace=True)

# Establecer la columna 'Date' como índice del DataFrame
poder_data.set_index('Date', inplace=True)

# Agrupar los datos por trimestre
trimestral_data = poder_data.resample('Q').mean()

# Crear las figuras y los ejes
fig, ax = plt.subplots(figsize=(15, 8))

# Graficar la media trimestral de 'Daily_Borb_Average'
ax.plot(trimestral_data.index, trimestral_data['Daily_Borb_Average'], label='Promedio Trimestral de Borb', marker='o')

# Graficar la media trimestral de 'InBorb'
ax.plot(trimestral_data.index, trimestral_data['InBorb'], label='Comportamiento Trimestral de InBorb', marker='x')

# Formatear el eje x con fechas
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Rotar las fechas para que sean legibles
plt.gcf().autofmt_xdate()

# Añadir leyendas y etiquetas
ax.set_xlabel('Fecha')
ax.set_ylabel('Valor Promedio')
ax.set_title('Comportamiento Trimestral de Borb e InBorb')
ax.legend()

# Mostrar el gráfico
plt.show()
```
