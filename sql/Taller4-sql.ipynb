```python
import sqlite3
import pandas as pd

# Conectar con la base de datos chinook.db
conn = sqlite3.connect('/content/chinook.db')

# Consulta 1: Selección Básica (Clientes en EE.UU. o Canadá)
query_usa_canada = """
SELECT FirstName, LastName, Email
FROM customers
WHERE Country = 'USA' OR Country = 'Canada'
"""
results_usa_canada = pd.read_sql_query(query_usa_canada, conn)
print("Clientes en EE.UU. o Canadá:")
print(results_usa_canada.head())
#results_usa_canada.to_csv('/mnt/data/customers_usa_canada.csv', index=False)

# Consulta 2: Unir Tablas (Lista de pistas para el género 'Rock')
query_rock_tracks = """
SELECT tracks.Name AS TrackName, albums.Title AS AlbumTitle, artists.Name AS ArtistName
FROM tracks
JOIN genres ON tracks.GenreId = genres.GenreId
JOIN albums ON tracks.AlbumId = albums.AlbumId
JOIN artists ON albums.ArtistId = artists.ArtistId
WHERE genres.Name = 'Rock'
"""
results_rock_tracks = pd.read_sql_query(query_rock_tracks, conn)
print("\nLista de pistas para el género 'Rock':")
print(results_rock_tracks.head())
#results_rock_tracks.to_csv('/mnt/data/rock_tracks_albums_artists.csv', index=False)

# Consulta 3: Ordenar y Limitar (10 facturas más caras)
query_expensive_invoices = """
SELECT customers.FirstName || ' ' || customers.LastName AS CustomerName,
       invoices.InvoiceDate AS InvoiceDate,
       invoices.Total AS Total
FROM invoices
JOIN customers ON invoices.CustomerId = customers.CustomerId
ORDER BY invoices.Total DESC
LIMIT 10
"""
results_expensive_invoices = pd.read_sql_query(query_expensive_invoices, conn)
print("\n10 facturas más caras:")
print(results_expensive_invoices.head())
#results_expensive_invoices.to_csv('/mnt/data/expensive_invoices.csv', index=False)

# Consulta 4: Agrupar y Contar (Lista de álbumes y conteo de pistas)
query_album_tracks_count = """
SELECT albums.Title AS AlbumTitle, COUNT(tracks.TrackId) AS TracksCount
FROM albums
JOIN tracks ON albums.AlbumId = tracks.AlbumId
GROUP BY albums.AlbumId
"""
results_album_tracks_count = pd.read_sql_query(query_album_tracks_count, conn)
print("\nLista de álbumes y conteo de pistas:")
print(results_album_tracks_count.head())
#results_album_tracks_count.to_csv('/mnt/data/album_tracks_count.csv', index=False)

# Cerrar la conexión con la base de datos
conn.close()
```
